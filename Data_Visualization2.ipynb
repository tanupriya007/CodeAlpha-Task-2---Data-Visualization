<h2 style="color:green;">Introduction</h2>

This task focuses on visualizing product review and sentiment data using Python. Various charts and plots have been created using Matplotlib and Seaborn to identify trends, patterns, and relationships within the dataset. The visualizations help in understanding product pricing, ratings, and customer sentiments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
%matplotlib inline
```

```python
df = pd.read_csv("Dataset-SA.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
print(df['product_price'].head())
print(df['Rate'].head())
```

```python
df['product_price'] = pd.to_numeric(df['product_price'], errors='coerce')
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
```

```python
df[['product_price', 'Rate']].dtypes
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['product_price'], bins=30)

plt.title('Distribution of Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Number of Products')

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Rate'], bins=5)

plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Sentiment', data=df)

plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')

plt.show()
```

```python
avg_rating = df.groupby('Sentiment')['Rate'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_rating, x='Sentiment', y='Rate')

plt.title('Average Rating by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Rating')

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='product_price',
    y='Rate'
)

plt.title('Product Price vs Rating')
plt.xlabel('Product Price')
plt.ylabel('Rating')

plt.show()
```

```python
avg_price = df.groupby('Sentiment')['product_price'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_price, x='Sentiment', y='product_price')

plt.title('Average Product Price by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Product Price')

plt.show()
```
